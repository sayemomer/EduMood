```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import os

# Example: Load all images from a directory
angry_image_path = os.listdir('./Dataset/images/train/angry')
disgust_image_path = os.listdir('./Dataset/images/train/disgust')
fear_image_path = os.listdir('./Dataset/images/train/fear')
happy_image_path = os.listdir('./Dataset/images/train/happy')
neutral_image_path = os.listdir('./Dataset/images/train/neutral')
sad_image_path = os.listdir('./Dataset/images/train/sad')
surprise_image_path = os.listdir('./Dataset/images/train/surprise')

# plot the number of images in each class
fig = plt.figure()
fig.suptitle('Number of images in each class')
ax = fig.add_axes([0,0,1,1])
emotion = ['angry', 'disgust', 'fear', 'happy', 'neutral', 'sad', 'surprise']
number_of_images = [len(angry_image_path), len(disgust_image_path), len(fear_image_path), len(happy_image_path), len(neutral_image_path), len(sad_image_path), len(surprise_image_path)]
ax.bar(emotion, number_of_images)
plt.show()

```

```python
# Present a collection of 25 images in a 5  5 grid,
# with the entire grid sized to fit within the confines of a standard letter-sized page.
# Ensure that the images are randomly chosen from each class upon every code execution. 
# This visual representation aids in understanding the datasets content and helps identify any noticeable anomalies or potential mislabelings.
# Path: image_eda.ipynb

import random

# randomly choose 25 images from each class
angry_image_path = random.sample(angry_image_path, 25)
disgust_image_path = random.sample(disgust_image_path, 25)
fear_image_path = random.sample(fear_image_path, 25)
happy_image_path = random.sample(happy_image_path, 25)
neutral_image_path = random.sample(neutral_image_path, 25)
sad_image_path = random.sample(sad_image_path, 25)
surprise_image_path = random.sample(surprise_image_path, 25)

# load images
angry_images = []
disgust_images = []
fear_images = []
happy_images = []
neutral_images = []
sad_images = []
surprise_images = []

image_classes = [angry_images, disgust_images, fear_images, happy_images, neutral_images, sad_images, surprise_images]

for i in range(25):
    angry_images.append(Image.open('./Dataset/images/train/angry/' + angry_image_path[i]))
    disgust_images.append(Image.open('./Dataset/images/train/disgust/' + disgust_image_path[i]))
    fear_images.append(Image.open('./Dataset/images/train/fear/' + fear_image_path[i]))
    happy_images.append(Image.open('./Dataset/images/train/happy/' + happy_image_path[i]))
    neutral_images.append(Image.open('./Dataset/images/train/neutral/' + neutral_image_path[i]))
    sad_images.append(Image.open('./Dataset/images/train/sad/' + sad_image_path[i]))
    surprise_images.append(Image.open('./Dataset/images/train/surprise/' + surprise_image_path[i]))

# plot images
fig = plt.figure(figsize=(10,10))
fig.suptitle('25 images from angry class')
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.imshow(angry_images[i], cmap='gray')
    plt.axis('off')
plt.show()


#25 random images  among all classes
random_image=[]
fig = plt.figure(figsize=(10,10))
fig.suptitle('25 random images')
for i in range(25):
    plt.subplot(5,5,i+1)
    ran_im = random.choice(random.choice(image_classes))
    plt.imshow(ran_im, cmap='gray')
    random_image.append(ran_im)
    plt.axis('off')
plt.show()



```
